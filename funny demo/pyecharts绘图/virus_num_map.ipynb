```python
from pyecharts.charts import Map
import pandas as pd
import numpy as np
from pyecharts import options as opts
```

```python
datapath = 'ZJ_virus_20200205.xlsx'
data = pd.read_excel(datapath)
data
```

```python
attr = list(data['area'])
value = data['num']
```

```python
def map_zhejiang() -> Map:
    c = (
        Map()
        .add("people_num", [list(z) for z in zip(attr, value)], "浙江")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-浙江地图"),
            visualmap_opts=opts.VisualMapOpts(),
        )
        .render('ZJ_virus_20200205.html')
    )
    return c
```

```python
c = (
        Map()
        .add("people_num", [list(z) for z in zip(attr, value)], "浙江")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="浙江地图"),
            visualmap_opts=opts.VisualMapOpts(),
        )
)
```

```python
c.render_notebook()
```

```python
data_CN = pd.read_excel('CN_virus_20200205.xlsx')
attr_CN = list(data_CN['area'])
value_CN = list(data_CN['num'])
```

```python
d = Map()
d.add('virus_people_num', [list(z) for z in zip(attr_CN, value_CN)],"china")
d.set_global_opts(title_opts=opts.TitleOpts(title="中国"),visualmap_opts=opts.VisualMapOpts())
d.render('CN_virus_20200205.html')
d.render_notebook()
```
