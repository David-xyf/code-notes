```python
from pyecharts.charts import Map
import pandas as pd
import numpy as np
from pyecharts import options as opts
```

```python
datapath = 'SX_virus_20201213.xlsx'
data = pd.read_excel(datapath)
data
```

```python
attr = list(data['area'])
value = data['population']
```

```python
def map_shanxi() -> Map:
    c = (
        Map()
        .add("people_num", [list(z) for z in zip(attr, value)], "山西")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-山西地图"),
            visualmap_opts=opts.VisualMapOpts(),
        )
        .render('SX_20201213.html')
    )
    return c
```
